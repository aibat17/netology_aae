```python
import pandas as pd
import re
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
```

```python
def traf(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return 'organic'
    elif row['region'] == 'Russia' and (row['traffic_source'] == 'paid' or row['traffic_source'] == 'email'):
        return 'ad'
    elif row['region'] != 'Russia' and (row['traffic_source'] == 'paid' or row['traffic_source'] == 'email'):
        return 'other'
    elif row['traffic_source'] not in ['yandex', 'google', 'paid','email']:
        return row['traffic_source']

log['source_type'] = log.apply(traf, axis = 1)
```

Задание 3 - прошу дать также обратную связь по более оптимизированному решению задачи. 

```python
films = pd.read_csv('ml-latest-small/ratings.csv')
films.head()
```

```python
films['date'] = pd.to_datetime(films['timestamp'], unit='s')

def diff(row):
    return row['max'] - row['min']

res = films.groupby('userId').agg(['count','min','max'])['date']
res_100 = res[res['count']> 100].reset_index()
res_100['diff'] = res_100.apply(diff, axis=1)
res_100.sort_values('diff', ascending=False)
```
