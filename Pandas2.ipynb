```python
import pandas as pd

data = pd.read_csv('ratings.csv')

def classif(row):
    if row['rating'] <= 2:
        return 'Низкий рейтинг'
    elif (row['rating'] > 2) & (row['rating'] <= 4):
        return 'Средний рейтинг'
    elif (row['rating'] == 4.5):
        return 'Высокий рейтинг'
    elif (row['rating'] == 5):
        return 'Высокий рейтинг'

data['class'] = data.apply(classif, axis=1)
data.head()
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад':['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

geo = pd.read_csv('keywords.csv')
geo.head()
#geo[geo['keyword'].str.contains('москва', case=False)]['keyword'].unique()

```

```python
def geo_class(col):
    for k,v in geo_data.items():
        for i in v:
            if i in col:
                return k
    return 'Undefined'


geo['regions'] = geo['keyword'].apply(geo_class)
```

```python
geo[geo['regions'] != 'Undefined']
```

```python
movie = pd.read_csv('movies.csv')
rating = pd.read_csv('ratings.csv')


merged = pd.merge(left=movie, right=rating, left_on='movieId',right_on='movieId', how='inner')
merged.head()
```

```python
years = list(range(1950, 2011))

def production_year(col):
    for i in years:
        if str(i) in col:
            return i
    return 1900

merged['year'] = merged['title'].apply(production_year)

merged.groupby('year').agg(['mean','max'])['rating'].reset_index().sort_values('year',ascending=False)
```
