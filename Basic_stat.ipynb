```python
import pandas as pd
import numpy as np
```

Задание 1

```python
horse = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv', sep=',',names=['surgery?', 'Age','Hospital number','rectal temperature','pulse','respiratory rate','temperature of extremities','peripheral pulse','mucous membranes','capillary refill time','pain','peristalsis','abdominal distension','nasogastric tube','nasogastric reflux','nasogastric reflux PH','rectal examination','abdomen','packed cell volume','total protein','abdominocentesis appearance','abdomcentesis total protein','outcome','surgical lesion?','lesion1','lesion2','lesion3','cp_data'])
```

```python
horse
```

```python
horse = horse[['surgery?','Age','rectal temperature','pulse','respiratory rate','temperature of extremities','pain','outcome']]
```

```python
#1
#тут я все вопросительные знаки заменил на nan так как при считывании dataframe 
#вопросительные знаки не распознавались как прпоуски, надеюсь это я сделал верно. 
#Если нет, то подскажите почему сразу автоматом не разпознались как пропуски?

horse = horse.replace('?',np.nan)
```

```python
horse.info()
```

```python
#2
#тут я перевед некоторые столбцы на тип float для возможности подсчета базовый статистики и также некоторые не стал ввиду того, 
#что они имеют Категориальный тип данных

horse['rectal temperature'] = horse['rectal temperature'].astype(float)
horse['pulse'] = horse['pulse'].astype(float)
horse['respiratory rate'] = horse['respiratory rate'].astype(float)
```

```python
horse.info()
```

Задание 2 

```python
#3
#Далее произвожу калькуляцию базовой статистики. 
#Видно что в каждом есть свой Максимум, выозможно является Выбросом,но столбец rectal temperature исключаю так как
#40 градусов это нормальное значение. проверяю дальше.... 
horse.describe() 
```

```python
#4
#Предполагаю, что Выбросы могут быть по вышеуказанным столбцам, посчитаю по каждому из Столбцов, 
#методом межинтервального размаха

q1 = horse['Age'].quantile(0.25)
q3 = horse['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['Age'].between(lower_bound,upper_bound,inclusive=True)].sort_values('Age')
remove_outliers
```

```python
#5
#получаем 24 значения с выбросом равным 9 по столбцу Age
outliers = pd.concat([horse, remove_outliers]).drop_duplicates(keep=False)
outliers.head()
```

```python
#6
#ВОПРОС - можно считать выбросы сразу по нескольким столбцам? 
```

```python
#7
#смотрю выбросы по столбцу pulse. Видим, что есть явно выделяющиеся значения такие как 164, 150,160,184, 150

q1 = horse['pulse'].quantile(0.25)
q3 = horse['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers1 = horse[horse['pulse'].between(lower_bound,upper_bound,inclusive=True)].sort_values('pulse')

outliers1 = pd.concat([horse, remove_outliers1]).drop_duplicates(keep=False)
outliers1.head()
```

```python
#8
#смотрю выбросы по столбцу respiratory rate. Видим, что есть явно значения не входящие в интервал = 8-10

q1 = horse['respiratory rate'].quantile(0.25)
q3 = horse['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers2 = horse[horse['respiratory rate'].between(lower_bound,upper_bound,inclusive=True)].sort_values('respiratory rate')

outliers2 = pd.concat([horse, remove_outliers2]).drop_duplicates(keep=False)
outliers2
```
