```python
#4
#Функция которая позвляет обросить выбросы с помощью входных параметров df и col
#НАДЕЮСЬ ТУТ Я ВАС ПРАВИЛЬНО ПОНЯЛ КАСАТЕЛЬНО ТОГО, ЧТОБЫ НАПИСАТЬ ФУНКЦИЮ

def count_outlier(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outliers = df[df[col].between(lower_bound,upper_bound,inclusive=True)].sort_values(col)
    return outliers
```

```python
count_outlier(horse,'pulse')
```

```python
#8
#Пропуски в столбце rectal temperature - исходя из вашей обратной связи, понимаю, что заполнять константой неверно. 
#Сделаем анализ. 
#ОСНОВЫВАЯСЬ НА ДАННЫХ surgery? И Age, РЕШАЮ ЗАПОЛНИТЬ ПРОПУСКИ СРЕДНИМ ЗНАЧЕНИЕМ, ТО ЕСТЬ surgery? = 1 И age = 1 
#ЗАПОЛНЯЮЮ СРЕДНИЕМ ЗНАЧЕНИЕМ 38.1 и т.д.

horse.groupby(['surgery?','Age'])['rectal temperature'].mean().round(1)
```

```python
#9
#ЗАПОЛНЯЮ ПРОПУСКИ В rectal temperature тут я обхожусь 4-ымя строками кода, так как НЕ ЗНАЮ МОЖНО ЛИ ОДНОЙ СТРОКОЙ ЗАПОЛНИТЬ 
#ПРОПУСКИ ПО 4-ЕМ УСЛОВИЯМ???

value1 = horse.loc[(horse['surgery?'] == 1) & (horse['Age'] == 1)]['rectal temperature'].median()
value2 = horse.loc[(horse['surgery?'] == 1) & (horse['Age'] == 9)]['rectal temperature'].median()
value3 = horse.loc[(horse['surgery?'] == 2) & (horse['Age'] == 1)]['rectal temperature'].median()
value4 = horse.loc[(horse['surgery?'] == 2) & (horse['Age'] == 9)]['rectal temperature'].median()

horse.loc[(horse['surgery?'] == 1) & (horse['Age'] == 1), 'rectal temperature'] = horse['rectal temperature'].fillna(value1)
horse.loc[(horse['surgery?'] == 1) & (horse['Age'] == 9), 'rectal temperature'] = horse['rectal temperature'].fillna(value2)
horse.loc[(horse['surgery?'] == 2) & (horse['Age'] == 1), 'rectal temperature'] = horse['rectal temperature'].fillna(value3)
horse.loc[(horse['surgery?'] == 2) & (horse['Age'] == 9), 'rectal temperature'] = horse['rectal temperature'].fillna(value4)

#ТУТ КСТАТИ Я НЕ ИСПОЛЬЗОВАЛ inplace=True ТАК КАК ЕСЛИ БУДУ ИСПОЛЬЗОВАТЬ ТО У МЕНЯ НЕОБОРОТ ВЕЗДЕ ПРОПИШЕТСЯ NaN
```

```python
#10
#Столбец pulse - предлагаю заполнить медианой 

horse['pulse'].fillna(horse['pulse'].median(), inplace=True)
horse.info()
```

```python
#11
#Столбец respiratory rate - предлагаю заполнить медианой.

horse['respiratory rate'].fillna(horse['respiratory rate'].median(), inplace=True)
horse.info()
```

```python
#12
#Столбец temperature of extremities - учитывая, что имеет категориальные данные считаю заполнить модой

horse['temperature of extremities'].value_counts()
```

```python
horse['temperature of extremities'].fillna(horse['temperature of extremities'].mode()[0], inplace=True)
horse.info()
```

```python
#13
#Столбец pain также категориальный, часть его пропусков (48 строк) можно заменить на значение 5 
#в случае если surgery = 1 и outcome = 2, остальные значения модой


horse.loc[(horse['surgery?'] == 1) & (horse['outcome'] == 2), 'pain'] = horse['pain'].fillna(5) #заполнение значением 5

horse['pain'].fillna(horse['pain'].mode()[0], inplace=True) #заполнение модой оставшихся
```
